```python
import tkinter as Tkinter
from tkinter import messagebox

class QuestionAppWindow(Tkinter.Tk):
    def __init__(self, questionList):
        Tkinter.Tk.__init__(self)
        self.initialize()
        self.questionNum = 0
        self.questionList = questionList
        self.Question=None

    def initialize(self):
        self.grid()
        self.columnconfigure(0, minsize=150)
        self.rowconfigure(0, minsize=20)
        self.rowconfigure(1, minsize=20)

        self.labStartvar = Tkinter.StringVar()
        self.labStart = Tkinter.Label(self, bg='white',
                                      textvariable=self.labStartvar, justify=Tkinter.LEFT)
        self.labStart.grid(row=0, column=0, sticky='NSEW')
        self.labStartvar.set(TextForLabelStart)

        self.butStartQuestion = Tkinter.Button(self, text='開始作答', command=self.onStartQuestion)
        self.butStartQuestion.grid(row=1, column=0, sticky='NSEW')

    def cleanAllWidget(self):
        for widget in self.grid_slaves():
            widget.grid_forget()

    def onStartQuestion(self):
        self.cleanAllWidget()
        self.questionList[self.questionNum].onDraw(self)

    def onNextQuestion(self,Question):
        self.Question = Question
        self.cleanAllWidget()
        self.questionNum = self.questionNum + 1
        if self.questionNum >= len(self.questionList):
            #All Question completed !!
            self.Question.onShowReport()
        else:
            self.questionList[self.questionNum].onDraw(self)


class Question():
    def __init__(self, ask, option1, o1value, option2, o2value):
        self.window = None
        self.ask = ask
        self.option1 = option1
        self.o1value = o1value
        self.option2 = option2
        self.o2value = o2value
        self.answervar = 0
        self.answervalue = 1000

    def onDraw(self, window):
        self.window = window
        window.grid()
        window.columnconfigure(0, minsize=150)
        window.rowconfigure(0, minsize=20)
        window.rowconfigure(1, minsize=20)
        window.rowconfigure(2, minsize=20)
        window.rowconfigure(3, minsize=20)

        labAsktvar = Tkinter.StringVar()
        labAsk = Tkinter.Label(window, bg='white',
                               textvariable=labAsktvar, wraplength=800)
        labAsk.grid(row=0, column=0, sticky='NSEW')
        labAsktvar.set(self.ask)

        self.answervar = Tkinter.IntVar()
        self.answervar.set(9000)
        o1 = Tkinter.Radiobutton(window, text=self.option1, variable=self.answervar, value=self.o1value)
        o1.grid(row=1, column=0, sticky='NW')
        o2 = Tkinter.Radiobutton(window, text=self.option2, variable=self.answervar, value=self.o2value)
        o2.grid(row=2, column=0, sticky='NW')

        if self.window.questionNum == len(self.window.questionList)-1:
            butNextQuestion = Tkinter.Button(window, text='完成作答', command=self.checkAnswer)
            butNextQuestion.grid(row=3, column=0)
        else:
            butNextQuestion = Tkinter.Button(window, text='下一題', command=self.checkAnswer)
            butNextQuestion.grid(row=3, column=0)
    def checkAnswer(self):
        if self.answervar.get() ==9000:
            Tkinter.messagebox.showerror(title='不行呦', message='你尚未作答本題。')
        else:
            a = self.answervar.get()
            self.answervalue = self.answervalue + a
            if self.window is not None:
                self.window.onNextQuestion(self)
    def onShowReport(self):
        self.window.grid()
        self.window.columnconfigure(0, minsize=150)
        self.window.rowconfigure(0, minsize=20)
        self.window.rowconfigure(1, minsize=20)
        self.window.rowconfigure(2, minsize=20)
        c = Q1.answervar.get()+Q2.answervar.get()+Q3.answervar.get()+Q4.answervar.get()+Q5.answervar.get()+Q6.answervar.get()+Q7.answervar.get()+Q8.answervar.get()+Q9.answervar.get()+Q10.answervar.get()+Q11.answervar.get()+Q12.answervar.get()
        finalscore = 1000 + c
        print(finalscore)

        if finalscore >=900:

            level = Tkinter.Label(self.window,bg='white', text='A')
            level.grid(row=0, column=0, sticky='NSEW')

            labExtvar = Tkinter.StringVar()
            levelexplain = Tkinter.Label(self.window,bg='white', textvariable=labExtvar, wraplength=800)
            levelexplain.grid(row=1, column=0, sticky='NSEW')
            labExtvar.set(Alevel)

        elif finalscore >=750:
            level = Tkinter.Label(self.window, bg='white', text='B')
            level.grid(row=0, column=0, sticky='NSEW')

            labExtvar = Tkinter.StringVar()
            levelexplain = Tkinter.Label(self.window, bg='white', textvariable=labExtvar, wraplength=800)
            levelexplain.grid(row=1, column=0, sticky='NSEW')
            labExtvar.set(Blevel)
        elif finalscore >=600:
            level = Tkinter.Label(self.window, bg='white', text='C')
            level.grid(row=0, column=0, sticky='NSEW')

            labExtvar = Tkinter.StringVar()
            levelexplain = Tkinter.Label(self.window, bg='white', textvariable=labExtvar, wraplength=800)
            levelexplain.grid(row=1, column=0, sticky='NSEW')
            labExtvar.set(Clevel)
        else:
            level = Tkinter.Label(self.window, bg='white', text='D')
            level.grid(row=0, column=0, sticky='NSEW')

            labExtvar = Tkinter.StringVar()
            levelexplain = Tkinter.Label(self.window, bg='white', textvariable=labExtvar, wraplength=800)
            levelexplain.grid(row=1, column=0, sticky='NSEW')
            labExtvar.set(Dlevel)

        labEndvar = Tkinter.StringVar()
        labEnd = Tkinter.Label(self.window, bg='white',
                           textvariable=labEndvar, wraplength=800, justify=Tkinter.LEFT)
        labEnd.grid(row=2, column=0, sticky='NSEW')
        labEndvar.set(EndTalk)


TextForLabelStart = "使用者你好^^"
```
